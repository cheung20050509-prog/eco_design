```python
"""
【示例】统计类函数
"""
import numpy as np

# 假设这是某只股票过去 5 天的收盘价
prices = np.array([100, 102, 101, 105, 107])

# 计算平均价格
mean_price = np.mean(prices)
print(f"平均收盘价: {mean_price}")

# 计算收盘价的中位数
median_price = np.median(prices)
print(f"收盘价中位数: {median_price}")

# 计算收盘价的标准差
std_price = np.std(prices)
print(f"收盘价标准差: {std_price}")

# 计算收盘价的方差
var_price = np.var(prices)
print(f"收盘价方差: {var_price}")

# 计算最低价和最高价
min_price = np.min(prices)
max_price = np.max(prices)
print(f"最低价: {min_price}, 最高价: {max_price}")

# 假设这是某只股票过去 5 天的成交量
volumes = np.array([1000, 1200, 1100, 1300, 1400])

# 计算总成交量
total_volume = np.sum(volumes)
print(f"总成交量: {total_volume}")

# 计算收盘价的 25% 和 75% 分位数
percentile_25 = np.percentile(prices, 25)
percentile_75 = np.percentile(prices, 75)
print(f"25% 分位数: {percentile_25}, 75% 分位数: {percentile_75}")

# 计算收盘价的直方图
hist, bin_edges = np.histogram(prices, bins=3)
print(f"直方图统计: {hist}")
print(f"区间边界: {bin_edges}")

# 假设这是另一只股票过去 5 天的收盘价
prices_2 = np.array([90, 92, 91, 95, 97])

# 计算两只股票价格的相关系数矩阵
corr_matrix = np.corrcoef(prices, prices_2)
print("相关系数矩阵:")
print(corr_matrix)

# 计算两只股票价格的协方差矩阵
cov_matrix = np.cov(prices, prices_2)
print("协方差矩阵:")
print(cov_matrix)
```

```python
"""
【示例】线性代数类函数
"""
import numpy as np

# 假设有3只股票的日收益率
returns = np.array([0.02, 0.01, -0.005])  # 3只股票的日收益率

# 假设我们投资这3只股票的权重分别是 0.5, 0.3, 0.2
weights = np.array([0.5, 0.3, 0.2])

# 计算投资组合的日收益率（加权和）
portfolio_return = np.dot(returns, weights)
print(f'投资组合的日收益率: {portfolio_return:.2%}')


# 假设有3只股票的每日收益率数据，每一列表示一只股票的收益
data = np.array([[0.02, 0.01, -0.005],
                 [0.03, 0.02, -0.002],
                 [-0.01, 0.005, 0.003]])

# 计算数据的转置
data_transposed = np.transpose(data)

print("转置后的收益数据：")
print(data_transposed)


# 假设我们有一个协方差矩阵
cov_matrix = np.array([[0.04, 0.01, 0.005],
                       [0.01, 0.03, 0.002],
                       [0.005, 0.002, 0.02]])

# 计算协方差矩阵的迹（即对角线元素的和）
trace_value = np.trace(cov_matrix)
print(f'协方差矩阵的迹: {trace_value}')


# 假设我们有一个协方差矩阵
cov_matrix = np.array([[0.04, 0.01],
                       [0.01, 0.03]])

# 计算协方差矩阵的行列式
det_value = np.linalg.det(cov_matrix)
print(f'协方差矩阵的行列式: {det_value}')


# 假设我们有一个2x2的协方差矩阵
cov_matrix = np.array([[0.04, 0.01],
                       [0.01, 0.03]])

# 计算协方差矩阵的逆
cov_matrix_inv = np.linalg.inv(cov_matrix)
print("协方差矩阵的逆：")
print(cov_matrix_inv)

# 假设我们有一个2x2的收益率数据矩阵
data = np.array([[0.02, 0.01],
                 [0.03, 0.02]])

# 计算奇异值分解
U, S, Vt = np.linalg.svd(data)

print("左奇异矩阵 U：")
print(U)
print("奇异值 S：")
print(S)
print("右奇异矩阵 Vt：")
print(Vt)

# 假设有一个表示股票回报的向量
returns = np.array([0.02, 0.03, -0.01])

# 计算该回报向量的欧几里得范数（即L2范数）
norm_value = np.linalg.norm(returns)
print(f'回报向量的L2范数: {norm_value}')
```

```python
"""
【示例37】蒙特卡罗模拟实验。
"""
# 设置股票价格的基本参数和模型
S0 = 100    # 初始股票价格
mu = 0.1    # 年化收益率
sigma = 0.2    # 年化波动率
T = 1    # 投资期限（年）
N = 252    # 投资期限内的交易日数
dt = T / N    # 时间步长

# 生成蒙特卡罗模拟实验的随机路径
M = 10000    # 模拟实验的路径数
S = np.zeros((M, N+1))    # 存储所有路径的股票价格
S[:, 0] = S0
for i in range(1, N+1):
    # 根据几何布朗运动模型生成每个时间步的股票价格
    S[:, i] = S[:, i-1] * np.exp((mu - 0.5*sigma**2)*dt + sigma*np.sqrt(dt)*np.random.normal(size=M))

# 计算每条路径的收益率
r = np.diff(S, axis=1) / S[:, :-1]

# 计算未来股票价格的期望值、方差、置信区间等统计特征
mean = np.mean(S[:, -1])
std = np.std(S[:, -1])
conf_int = np.percentile(S[:, -1], [2.5, 97.5])

# 输出结果
print("期望股票价格：", mean)
print("标准差：", std)
print("95%置信区间：", conf_int)
```
